```python
y = filteredRating.groupby('Book-Title').count()['Book-Rating'] > 50
y[y].index
famous_books = y[y].index
```

```python
filteredRating[filteredRating['Book-Title'].isin(famous_books)]
```

```python
finalRatings =filteredRating[filteredRating['Book-Title'].isin(famous_books)]
```

```python
pt = finalRatings.pivot_table(index='Book-Title',columns='User-ID', values = 'Book-Rating')
```

```python
pt
```

```python
pt.shape
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt)
```

```python
score = cosine_similarity(pt)
```

```python
score
```

```python
sorted(list(enumerate(score[0])), key=lambda x:x[1], reverse=True)[1:6]
```

```python
def recommend(book_name):
  index = np.where(pt.index==book_name)[0][0]
  distances = score[index]
  similar_books = sorted(list(enumerate(score[index])), key=lambda x:x[1], reverse=True)[1:6]

  for i in similar_books:
    print(pt.index[i[0]])
```
